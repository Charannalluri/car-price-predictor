#converted Km_driven , mileage columns in to numeric

```python
df['km_driven'] = pd.to_numeric(df['km_driven'], errors = "coerce")
```

```python
df['mileage'] = pd.to_numeric(df['mileage'], errors = "coerce")
```

drop the Unnamed:0 column

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.info()
```

#Task 2:

```python
df.isnull().sum()
```

```python
#Dropping null values from target column
df = df.dropna(subset=['selling_price'])
```

```python
df.isnull().sum()
```

```python
#Impute missing input features
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

```python
#For categorical features with mode
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode())
```

```python
df.isnull().sum()
```

Strip and Lowercase the "Brand" Column

```python
df['brand'] = df['brand'].str.strip().str.lower()
```

```python
df.head()
```

Extract Numeric Values from "Mileage"

```python
df['mileage'] = df['mileage'].str.replace(',',  '').str.extract('(\d+)'.astype(float))
```

```python
print(df['mileage'].dtype)
```

```python
#remove duplicates
df = df.drop_duplicates()
```

#Task 3:

```python
#Calculate the Mean Selling Price
y_mean = df['selling_price'].mean()
print(y_mean)
```

```python
#Create Baseline Predictions
y_pred_baseline = [y_mean] * len(df)
print(y_pred_baseline)
```

```python
#Calculate and Print MAE
from sklearn.metrics import mean_absolute_error

#calculate mae
baseline_mae = mean_absolute_error(df['selling_price'], y_pred_baseline)
```

```python
print(f"Mean selling_price: {y_mean:.2f}")
print(f"Baseline MAE: {baseline_mae:.2f}")
```
